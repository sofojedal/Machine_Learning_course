```python
np.random.seed(123)
x = np.arange(0, 5, 0.1)
y = 3 * x  +  2  + np.random.normal(0, 4, len(x))
```

```python
plt.scatter(x, y)
```

1. Encontrar una línea 1 que pase por los puntos

```python
from sklearn.linear_model import LinearRegression

regression1 = LinearRegression()
regression1.fit(x.reshape(-1, 1), y)
y_pred1 = regression1.predict(x.reshape(-1, 1))
y_pred1
```

2. Encontrar una línea 2 que pase por los puntos

```python
regression2 = LinearRegression()
regression2.fit(x[10:].reshape(-1, 1), y[10:])
y_pred2 = regression2.predict(x.reshape(-1, 1))
y_pred2
```

3. Dibujar las líneas en el scatter

```python
plt.scatter(x, y, label="Datos reales")
plt.plot(x, y_pred1, label="Regresión Lineal 1")
plt.plot(x, y_pred2, label="Regresión Lineal 2")
```

4. Determinar cuál tiene mejor MSE

```python
mse_linear1 = mean_squared_error(y, y_pred1)
mse_linear2 = mean_squared_error(y, y_pred2)

if mse_linear1 < mse_linear2:
    print("La Regresión Lineal 1 tiene un MSE más bajo.")
else:
    print("La Regresión Lineal 2 tiene un MSE más bajo.")

print(mean_squared_error(y, y_pred1))
print(mean_squared_error(y, y_pred2))
```

5. Encontrar un árbol regresor de altura 3

```python
tree_reg1 = DecisionTreeRegressor(max_depth=3)
tree_reg1.fit(x.reshape(-1, 1), y)
y_tree_pred1 = tree_reg1.predict(x.reshape(-1, 1))
```

6. Encontrar un árbol regresor sin altura máxima

```python
tree_reg2 = DecisionTreeRegressor()
tree_reg2.fit(x.reshape(-1, 1), y)
y_tree_pred2 = tree_reg2.predict(x.reshape(-1, 1))
```

7. KNN regressor con k = 2

```python
knn_reg2 = KNeighborsRegressor(n_neighbors=2)
knn_reg2.fit(x.reshape(-1, 1), y)
y_knn_pred2 = knn_reg2.predict(x.reshape(-1, 1))
```

8. KNN regressor con k = 3

```python
knn_reg3 = KNeighborsRegressor(n_neighbors=3)
knn_reg3.fit(x.reshape(-1, 1), y)
y_knn_pred3 = knn_reg3.predict(x.reshape(-1, 1))
```

9. Encontrar el MSE para cada modelo

```python
mse_tree1 = mean_squared_error(y, y_tree_pred1)
mse_tree2 = mean_squared_error(y, y_tree_pred2)
mse_knn2 = mean_squared_error(y, y_knn_pred2)
mse_knn3 = mean_squared_error(y, y_knn_pred3)
```

10. Dibujar x vs y para cada modelo sobre el scatter

```python
plt.scatter(x, y, label="Datos reales")
plt.plot(x, y_pred1, label="Regresión Lineal 1")
plt.plot(x, y_pred2, label="Regresión Lineal 2")
plt.plot(x, y_tree_pred1, label="Árbol Regresor (Altura 3)")
plt.plot(x, y_tree_pred2, label="Árbol Regresor (Sin Altura Máxima)")
plt.plot(x, y_knn_pred2, label="KNN Regresor (k=2)")
plt.plot(x, y_knn_pred3, label="KNN Regresor (k=3)")
```

```python
# pendiente
m = 0
best_m = None
best_mse = float('inf')
m_values = []
mse_values = []


# Bucle para calcular el MSE para diferentes valores de pendiente
while m <= 7:
    y_pred = m * x
    mse = mean_squared_error(y, y_pred)
    # Almacenamos los valores de m y MSE
    m_values.append(m)
    mse_values.append(mse)
# Comparamos con el mejor MSE hasta ahora
    if mse < best_mse:
        best_mse = mse
        best_m = m
    m += 0.1

# Imprimimos el mejor resultado
print(f"Mejor pendiente (m): {best_m}, Mejor MSE: {best_mse}")
```

```python
best_m = None
best_b = None
best_mse = float('inf')

# Bucle para calcular el MSE para diferentes valores de m y b
for m in np.arange(0, 7, 0.1):
    for b in np.arange(0, 7, 0.1):
        y_pred = m * x + b
        mse = mean_squared_error(y, y_pred)
        # Comparamos el MSE actual con el mejor MSE encontrado hasta ahora
        if mse < best_mse:
            best_mse = mse
            best_m = m
            best_b = b

# Imprimimos los mejores valores
print(f"Mejor Pendiente (m): {best_m}")
print(f"Mejor coordenada al Origen (b): {best_b}")
print(f"Mejor MSE: {best_mse}")
```

```python
plt.figure(figsize=(10, 6))
plt.plot(m_values, mse_values, marker='o')
plt.title('Relación entre Pendiente (m) y MSE')
plt.xlabel('Pendiente (m)')
plt.ylabel('MSE')
plt.grid(True)
plt.show()
```
