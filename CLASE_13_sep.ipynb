```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
```

```python
np.random.seed(123)
x = np.arange(0, 5, 0.1)
y = 3 * x  +  2  + np.random.normal(0, 4, len(x))
```

Bucle para calcular el MSE para diferentes valores  de pendiente, en este while se almacenan los valores de m y MSE, luego comparamos con el mejor MSE hasta ahora; y por último imprimimos el mejor resultado

```python
m = 0
best_m = None
best_mse = float('inf')
m_values = []
mse_values = []

while m <= 7:
    y_pred = m * x
    mse = mean_squared_error(y, y_pred)

    m_values.append(m)
    mse_values.append(mse)

    if mse < best_mse:
        best_mse = mse
        best_m = m
    m += 0.1

print(f"Mejor pendiente (m): {best_m}, Mejor MSE: {best_mse}")
```

Bucle para calcular el MSE para diferentes valores de m y b, donde en el if, comparamos el MSE con el mejor MSE encontrado hasta ahora, al final imprimimos los mejores valores

```python
best_m = None
best_b = None
best_mse = float('inf')

for m in np.arange(0, 7, 0.1):
    for b in np.arange(0, 7, 0.1):
        y_pred = m * x + b
        mse = mean_squared_error(y, y_pred)

        if mse < best_mse:
            best_mse = mse
            best_m = m
            best_b = b

print(f"Mejor Pendiente (m): {best_m}")
print(f"Mejor coordenada al Origen (b): {best_b}")
print(f"Mejor MSE: {best_mse}")
```

A continuación se muestra la gráfica bidimensional

```python
plt.figure(figsize=(10, 6))
plt.plot(m_values, mse_values, marker='o')
plt.title('Relación entre Pendiente (m) y MSE')
plt.xlabel('Pendiente (m)')
plt.ylabel('MSE')
plt.grid(True)
plt.show()
```
