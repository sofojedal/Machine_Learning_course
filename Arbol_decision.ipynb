```python
  from sklearn.tree import plot_tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
x_train = [[13], [15], [17], [20], [23], [25]]
y_train = ["NO PASA", "PASA", "PASA", "NO PASA", "NO PASA", "NO PASA"]
```

```python
x_test = [[14], [25], [16]]
y_test = ["PASA", "NO PASA", "PASA"]
```

```python
model = DecisionTreeClassifier()
```

```python
# Entrenar a la máquina
model.fit(x_train,y_train)
```

```python
plot_tree(model)
```

```python
accuracy_score(y_test, model.predict(x_test))*100
```

```python
model = DecisionTreeClassifier(max_depth=1)
```

```python
model.fit(x_train,y_train)
```

```python
plot_tree(model)
```

```python
accuracy_score(y_test, model.predict(x_test))*100
```

```python
accuracy_score(y_train, model.predict(x_train))*100
```
